```python
lote_buscado = 8304020002
```

```python
df_lote_buscado = df[df['LOTCODIGO'] == lote_buscado]
```

```python
import matplotlib.pyplot as plt

# Filter the dataframe based on conditions
filtered_data = df_lote_buscado
cuantia_values = filtered_data['cuantia']

# Plotting histogram
plt.hist(cuantia_values, bins=50, color='blue', edgecolor='black')

# Calculate mean and median
min_value = cuantia_values.min()
max_value = cuantia_values.max()
mean_value = cuantia_values.mean()
median_value = cuantia_values.median()

# Plotting mean and median as vertical lines
# plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=1, label='Mean')
# plt.axvline(median_value, color='green', linestyle='dashed', linewidth=1, label='Median')

# Adding text annotations for mean and median values
# plt.text(mean_value, 80, f'Mean: {mean_value:.2f}', rotation=0, color='red', fontsize=10, va='bottom', ha='left')
# plt.text(median_value, 50, f'Median: {median_value:.2f}', rotation=00, color='green', fontsize=10, va='bottom', ha='left')

# Set x-axis limit
plt.xlim(min_value, max_value) 
 
# Adding labels and title
plt.xlabel('Cuantia')
plt.ylabel('Frequency')
plt.title('Histogram of Cuantia')

# Adding legend
plt.legend()

# Display the plot
plt.show()

min_value = locale.currency(min_value, grouping=True)
mean_value = locale.currency(mean_value, grouping=True)
median_value = locale.currency(median_value, grouping=True)
max_value = locale.currency(max_value, grouping=True)
```

```python
print(f'Min {min_value}')
print(f'Mean {mean_value}')
print(f'Median {median_value}')
print(f'Max {max_value}')
```
